```python
# Importando las bibliotecas necesarias
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn import datasets
from sklearn.metrics import accuracy_score

# Cargando el conjunto de datos 'Digits'
digits = datasets.load_digits()
X = digits.data
y = digits.target

# Descripcion del dataset
digits_description = digits.DESCR
print(digits_description)




```

```python
# numero de casos o instancias
num_cases = digits.data.shape[0]
print(num_cases)
```

```python

# Obteniendo los valores únicos de la clase en el conjunto de datos 'Digits'
class_values = set(digits.target)
print(class_values)
```

```python
# Importando la biblioteca matplotlib
import matplotlib.pyplot as plt
import seaborn as sns

# Creando la gráfica
plt.figure(figsize=(10, 6))
sns.countplot(x=digits.target)
plt.title("Distribución de Clases en el Conjunto de Datos 'Digits'")
plt.xlabel("Clase (Dígito)")
plt.ylabel("Frecuencia")
plt.xticks(range(10))
plt.show()
```

```python
# Creando un histograma de la quinta variable (quinto píxel) del conjunto de datos 'Digits'
plt.figure(figsize=(10, 6))
sns.histplot(digits.data[:, 4], kde=True, bins=16)  # Indice 4 para la quinta variable
plt.title("Histograma del Quinto Píxel en el Conjunto de Datos 'Digits'")
plt.xlabel("Valor del Píxel")
plt.ylabel("Frecuencia")
plt.show()
```

```python
# Dividiendo los datos en un conjunto de entrenamiento y un conjunto de prueba (70-30)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Inicializando el clasificador Naive Bayes
gnb = GaussianNB()

# Entrenando el modelo con el conjunto de entrenamiento
gnb.fit(X_train, y_train)

# Realizando predicciones en el conjunto de prueba
y_pred = gnb.predict(X_test)

# Calculando y mostrando la precisión del modelo
accuracy = accuracy_score(y_test, y_pred)
accuracy
```

```python
from sklearn.metrics import classification_report

# Calculando y mostrando F1-score, recall, precisión y otras métricas
metrics_report = classification_report(y_test, y_pred)
print(metrics_report)
```
